```python
import requests
from bs4 import BeautifulSoup
```

```python
url1 = "https://codewithharry.com"
url2 = "https://www.google.com"
```

# Step-1 Get the HTML

```python
r = requests.get(url1)
htmlContent  = r.content
print(htmlContent)
```

# Step-2 Parse the HTML

```python
soup = BeautifulSoup(htmlContent)
print(soup.prettify)
```

# Html Tree Traversal

```python
title = soup.title
print(title.string)
```

```python
# 4 objects last one is Comment
print(type(soup))
print(type(title))
print(type(title.string))
```

```python
paras = soup.find_all("p")   # Find all paras 
anchors = soup.find_all("a")
div = soup.find_all("div")
print(paras,"\n")
print(anchors,"\n")
print(div)
```

```python
print(soup.find("p")["class"])        # Gives all the class related to the paragraph tag 
```

```python
print(soup.find_all("p",class_="lead"))    # find all paragraph tag having the class lead
```

```python
print(soup.find("p").get_text())
```

```python
print(soup.get_text())        # returns the text of the website
```

```python
allLinks = set()
link_text = "https://codewithharry.com"
for links in anchors:
    if(links.get("href")!="#"):
        allLinks.add(link_text+links.get("href"))
        
print(allLinks)        
```

```python
markup = "<p><!-- Hello Nikhil Rana is Here --></p>"
soup2 = BeautifulSoup(markup)

print(type(soup2.p.string))
```

```python
# Content: A tag's children are available as list
# children: A tag's children are avaliable as generator

navContent = soup.find(id = "navbarSupportedContent")
for elem in navContent.contents:
    print(elem)
```

```python
for x in navContent.stripped_strings:
    print(x)                                  # Strings and stripped_strings
```

```python
for x in navContent.parents:                         # parents AND parent
    print(x.name)
```

```python
for y in navContent.previous_sibling.previous_sibling:
    print(y)
```

```python
for y in navContent.next_sibling:
    print(y)
```
